# Pre-Processing of Data


```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
data = data.rename( columns= {
                          'v1' : 'tag' ,
                          'v2' : 'text'
                      }
           )

```

```python
data
```

# Text to Numerical

```python
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(data['text'])
```

# Prepare Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(features, data['tag'], test_size=0.2, random_state=42)
print(features)
```

```python
print(X_train)
```

# Create Model

```python
 # Training Naive Baises Classifier
model = MultinomialNB()
model.fit(X_train, y_train)
```

# Evaluation of model

```python
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
new_df = pd.DataFrame(
    {
        'Text' : X_test,
        'Predicted_tag' : predictions ,
        'Expected_tag' : y_test
    }
)
```
